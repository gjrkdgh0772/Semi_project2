# X1~X5 까지의 변수들 자기자본만 자산총계-부채총계 하면된다.

```python
import urllib.request
from bs4 import BeautifulSoup
종목_list = []
매출액_totlist = []
유동자산_totlist = []
유동부채_totlist = []
총자산_list = []
총부채_list = []
이익잉여금_list = []
for index in kospi_list2018:

    try:
    
        # 네이버 금융 국내증시 메인 사이트 주소
        url = "https://comp.fnguide.com/SVO2/ASP/SVD_Finance.asp?pGB=1&gicode=A"+index+"&cID=&MenuYn=Y&ReportGB=&NewMenuID=103&stkGb=701"
        # 웹사이트 정보 요청
        page = urllib.request.urlopen(url)

        # 해당 페이지는 cp949 방식의 인코딩 사용
        html = page.read().decode('utf8')

        # html.parser로 html 형식 태그에서 데이터 추출 준비

        soup = BeautifulSoup(html, 'html.parser')
        종목이름 = soup.select_one('#giName').text
        종목_list.append(종목이름)

        매출액 =  soup.find(id='divSonikY').find_all('tr')[1].find_all('td')
        매출액_list = []
        for number in 매출액:
            매출액_list.append(number.string)
        매출액_totlist.append(매출액_list)



        유동자산 = soup.find(id='divDaechaY').find_all('tr')[2].find_all('td')
        유동자산_list = []
        for number in 유동자산:
            유동자산_list.append(number.string)
        
        유동자산_totlist.append(유동자산_list)

        유동부채 = soup.find(id='divDaechaY').find_all('tr')[30].find_all('td')
        유동부채_list = []
        for number in 유동부채:
            유동부채_list.append(number.string)
        유동부채_totlist.append(유동부채_list)

        자산 =  soup.find(id='divDaechaY').find_all('tr')[1].find_all('td')
        자산_list = []
        for number in 자산:
            자산_list.append(number.string)
        총자산_list.append(자산_list)

        부채 =  soup.find(id='divDaechaY').find_all('tr')[29].find_all('td')
        부채_list = []
        for number in 부채:
            부채_list.append(number.string)
        총부채_list.append(부채_list)

        이익잉여금 = soup.find(id='divDaechaY').find_all('tr')[65].find_all('td')
        이익_list = []
        for number in 이익잉여금:
            이익_list.append(number.string)
        이익잉여금_list.append(이익_list)


    except:
        pass
print(종목_list)
print(매출액_totlist)
print(유동자산_totlist)
print(유동부채_totlist)
print(총자산_list)
print(총부채_list)
print(이익잉여금_list)
```

```python
Z = pd.DataFrame()
```

```python
Z.index = 종목_list
Z['매출액_totlist'] = 매출액_totlist
Z['유동자산_totlist'] = 유동자산_totlist
Z['유동부채_totlist'] = 유동부채_totlist
Z['총자산_list'] = 총자산_list
Z['총부채_list'] = 총부채_list
Z.to_csv('Z',index='False')
```

# ROA 생성기 이건 2017년 부터 2021년까지 5개까지 끌어서 온다

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from tqdm import tqdm
from time import sleep

ROA_list = []
종목_list = []
for index in tqdm(kospi_list2018, desc='tqdm test processing'):

    try:
        

        url = "https://comp.fnguide.com/SVO2/ASP/SVD_FinanceRatio.asp?pGB=1&gicode=A"+index+"&cID=&MenuYn=Y&ReportGB=&NewMenuID=104&stkGb=701"
        html = urlopen(url).read()
        soup = BeautifulSoup(html, 'html.parser')

        ROA = soup.find(id='compBody').find_all('tr')[52].find_all('td')
        roa_list = []
        for number in ROA:

            
            roa_list.append(number.string)
        ROA_list.append(roa_list)

        종목이름 = soup.select_one('#giName').text
        종목_list.append(종목이름)
    except:
        pass
print(종목_list)
print(ROA_list)
# for i in tmp:
#     print(i.string)


```
